```python
#create client to bigquery
client = bigquery.Client()
```

```python
#define the query

QUERY= (
'SELECT * FROM `iron-potion-771.Res.RawJson` LIMIT 5'
)
```

```python
#Create lists for each data element:   transaction type and vendor code

transType = []
vendorCode = []
```

```python
#iterate through each row, and parse the specific data requested
#Reference:  https://stackoverflow.com/questions/46186733/instagramapi-how-to-parse-json-python-data-into-pandas-dataframe-ipython-jup

for result in apple['results']:
    transType.append(result['metadata']['type'])
    vendorCode.append(result['message']['reservations']['vendorCode'])
```

```python
#Create a dataframe with the parsed data 

parsedData = pd.DataFrame([transType,vendorCode]).T
parsedData.head()
```
